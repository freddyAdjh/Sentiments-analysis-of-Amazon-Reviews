```python
import pandas as pd
import numpy as np
import re
import nltk
import string
string.punctuation
from nltk.stem import WordNetLemmatizer
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
en_stopwords = nltk.corpus.stopwords.words('english')
wordLem = WordNetLemmatizer()
```

```python
Amazon_review = pd.read_csv('./amazon.csv')
Amazon_review = Amazon_review.iloc[:,1:]
Amazon_review.head(5)
```

```python
Amazon_review['overall'].value_counts(ascending=True).plot.barh()
```

```python
Amazon_review.shape
```

```python
Amazon_review = Amazon_review.dropna()
```

```python
Amazon_review.shape
```

```python
Amazon_review[Amazon_review['overall']==1].iloc[1,:]['reviewText']
```

```python
def assign_class(reviews):
    if int(reviews)==1 or int(reviews)==2:
        return 0
    elif int(reviews)==5 or int(reviews)==4:
        return 2
    else:
        return 1
    
```

```python
%%time
Amazon_review['Sentiment'] = Amazon_review['overall'].apply(lambda x:assign_class(x))
```

```python
Amazon_review.columns
```

```python
Dataset = Amazon_review.loc[:,['Sentiment','reviewText']]
Dataset.sample(5)
```

#### 1 : Neutral
**2: Positive**
**0: Negative**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.countplot(Dataset,x="Sentiment")
plt.show()
```

```python
def clean_reviews(reviews):
    reviews = "".join([word.lower() for word in reviews if word not in string.punctuation])
    reviews = re.split("\W+",reviews)
    reviews = [wordLem.lemmatize(word) for word in reviews if word not in en_stopwords]
    
    reviews = " ".join(reviews)
    
    return reviews
```

```python
%%time
Dataset['reviews_cleaned'] = Dataset.reviewText.apply(lambda x: clean_reviews(x))
```

```python
Dataset.head(5)
```

```python
wc = WordCloud(background_color='white',max_font_size=100,max_words=70)
```

```python
def reviewsCloud(df):
    return " ".join([reviews for reviews in df['reviews_cleaned']])
```

```python
def plotWordCloud(wordCloud):
    plt.figure(figsize=(10,8))
    plt.imshow(wc.generate(wordCloud))
    plt.title("reviews words cloud")
    plt.show()
```

## All reviews

```python
allReviews = reviewsCloud(Dataset)
plotWordCloud(allReviews)
```

**We can see that most of the reviews in the dataset are about storage**

## Postive reviews words cloud

## Neutral reviews words cloud

```python
neutralReviews = reviewsCloud(Dataset[Dataset['Sentiment']==1])
plotWordCloud(neutralReviews)
```

### Vectorization using tf-idf technique

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
vectorizer = CountVectorizer(max_features=1000)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X = vectorizer.fit_transform(Dataset['reviews_cleaned']).toarray()
Y = Dataset['Sentiment'].values
```

```python
X.shape
```

```python
X.min()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train,X_val,Y_train,Y_val = train_test_split(X,Y,random_state=42,stratify=Y)
```

```python
X_train.max(),X_val.max()
```

```python
X_train.shape,X_val.shape
```

```python
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
X_train.max(),X_val.max()
```

## Modelization

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
logReg_model = LogisticRegression()
KNN_model = KNeighborsClassifier()
svm_model = SVC()
RndFor_model = RandomForestClassifier()
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
models = {
    'logistic Regression':logReg_model,
    'KNN':KNN_model,
    'SVM':svm_model,
    'Random Forest':RndFor_model
}
```

```python
%%time
for k,model in models.items():
    print(f"Model {k}".center(80,'~'))
    print(f"Accuracy: {cross_val_score(model,X,Y,cv=15,scoring='accuracy').mean()}")
```

**As we can see, logistic regression model is the best model option for our task. So are going now to find its bests parameters**

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,f1_score,accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': np.logspace(-5, 5, 50),
             'penalty':['l1', 'l2']}
clf = GridSearchCV(LogisticRegression(random_state=42), param_grid,cv=5, verbose=0,n_jobs=-1)
```

```python
%%time
model = clf.fit(X_train,Y_train)
print(model.best_estimator_)
```

```python
%%time
logReg_model = LogisticRegression(C=1.2648552168552958, random_state=42)
logReg_model.fit(X_train, Y_train)
y_pred = logReg_model.predict(X_val)

cm = confusion_matrix(y_pred,Y_val)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=logReg_model.classes_)
cm_display.plot()
plt.show()
```

```python
print(f"Accuracy: {logReg_model.score(X_val,Y_val)*100:.2f} %")
```
